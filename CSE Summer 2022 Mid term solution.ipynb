```python
class Exam:
    def __init__(self,ex,ques,marks):
        self.ex=ex
        self.ques=ques
        self.marks=marks
    def detail(self):
        print(f" Exam Type : {self.ex} \n Number Of Questions : {self.ques} \n Marks Per Question : {self.marks} ")
        if self.ex == "Midterm":
            print(" Total Marks:" , 2*self.marks)
        else :
            print(" Total Marks:" , 3*self.marks)

e1 = Exam('Midterm', 2, 10)
print(e1.detail())
print("===========================")
e2 = Exam('Final', 3, 10)
print(e2.detail())
```

```python
class RickMote:
    def __init__(self):
        self.pw = "OFF"
        self.channel = 0
        self.volume = 3
        self.channellist = [0,2,3,6,7,9]
        self.i = 0 
    
    def power(self):
        if self.pw=="OFF":
            self.pw="ON"
        else:
            self.pw="OFF"
    def changeChannel(self,ch=None):
        if self.pw=="ON":
            if ch==None:
                self.channel = self.channellist[self.i]
                self.i+=1
            elif ch not in self.channellist:
                print("TV channel does not exist.")
            else:  
                self.channel = ch
        else:
            print("power is turned off. Cannot change channel.")
    
    def changeVolumeLevel(self,vl=1):
        if self.pw=="ON":
            self.volume += vl
        else:
            print("power is turned off. Cannot change volume.")
    
    def showInfo(self):
        if self.pw=="OFF":
            print("ID Cable Box Status:\nCable Box is: OFF")
        else:
            print(f"ID Cable Box Status:\nCable Box is: ON\nChannel:{self.channel}\nVolume:{self.volume}")


oTV = RickMote()
oTV.power()
print("1.############################")
oTV.showInfo()
print("2.############################")
oTV.changeChannel()
oTV.changeVolumeLevel()
oTV.showInfo()
print("3.############################")
oTV.power()
oTV.showInfo()
print("4.############################")
oTV.power()
oTV.changeVolumeLevel(4)
oTV.changeChannel(3)
oTV.showInfo()
print("5.############################")
oTV.changeVolumeLevel(-2)
oTV.showInfo()
print("6.############################")
oTV.power()
oTV.changeChannel(9)
oTV.changeVolumeLevel(-1)
oTV.showInfo()
print("7.############################")
oTV.power()
oTV.changeChannel(11)
oTV.showInfo()
```
